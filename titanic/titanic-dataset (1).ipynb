3) What was the overall survival rate?

```python
import matplotlib.pyplot as plt
total_survived=data.groupby('Survived').count()['PassengerId'][1]
not_survived=data.groupby('Survived').count()['PassengerId'][0]
total=data['Survived'].count()
survival_rate=(total_survived/total)*100
not_survival_rate =(not_survived/total)*100
plt.pie([total_survived,not_survived],labels=["Survived", "Not Survived"],autopct='%1.1f%%')

print(f'The overall survival rate is {survival_rate :.2f}%.')
print(f'The overall survival rate is {not_survival_rate :.2f}%.')
```
