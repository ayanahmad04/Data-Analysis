1) What percentage of passengers are satisfied vs. dissatisfied?



```python
import matplotlib.pyplot as plt
data.columns
data['Satisfied']
total = data['Satisfied'].count()
satisfied =  data.groupby('Satisfied')['Ref'].count()[1]
not_satisfied =  data.groupby('Satisfied')['Ref'].count()[0]
plt.pie([satisfied ,not_satisfied],labels=["Satisfied", "Not Satisfied"],autopct='%1.1f%%')
# data.groupby('Satisfied')['Ref'].count().plot.bar()
data.groupby('Satisfied')['Ref'].count()
```

```python
data.columns
data.groupby('Gender')
satisfied=data[data['Satisfied']=='Y'].groupby('Gender')['Ref'].count()
not_satisfied=data[data['Satisfied']=='N'].groupby('Gender')['Ref'].count()
df_plot = pd.DataFrame({'Satisfied': satisfied, 'Unsatisfied': not_satisfied})
df_plot.plot.bar()
```

```python
plt.figure(figsize=(15, 3)) 
plt.subplot(1,1,1)
plt.pie([satisfied[0],not_satisfied[0]],labels=["Satisfied", "Not Satisfied"],autopct='%1.1f%%',)
plt.title('Female')
plt.subplot(1,2,2)
plt.pie([satisfied[1],not_satisfied[1]],labels=["Satisfied", "Not Satisfied"],autopct='%1.1f%%',)
plt.title('Male')
plt.tight_layout()
```

```python
data.columns
data['Type of Travel'].unique()
satisfiedd=data[data['Satisfied']=='Y'].groupby('Type of Travel')['Ref'].count()
not_satisfiedd=data[data['Satisfied']=='N'].groupby('Type of Travel')['Ref'].count()
# df=pd.DataFrame({'satisfied':satisfied,'not_satisfied':not_satisfied})
# df.plot.bar()
plt.figure(figsize=(10, 3)) 
plt.subplot(1,2,1)
plt.pie([satisfiedd[0],not_satisfiedd[0]],labels=['satisfied','not_satisfied'],autopct='%1.1f%%')
plt.title('Business Class')
plt.subplot(1,2,2)
plt.pie([satisfiedd[1],not_satisfiedd[1]],labels=['satisfied','not_satisfied'],autopct='%1.1f%%')
plt.title('Personal Class')
```

3) What age groups are most represented in the dataset?

```python
import matplotlib.pyplot as plt
data['Age Band'].value_counts()
# data.columns
# data['Age Band'].isna().sum()
plt.figure(figsize=(10, 6)) 
plt.subplot(1,2,1)
data.groupby('Age Band')['Ref'].count().plot.bar()
band=data.groupby('Age Band')['Ref'].count()
# len(band)
plt.subplot(1,2,2)
labels=['18 to 24','25 to 34','35 to 44','45 to 54','55 to 64','65 or over','Under 18']
plt.pie([x for x in band],labels=labels,autopct='%1.1f%%')
```
